```python
import numpy as np 
import matplotlib.pyplot as plt
```

#### Cálculo de histograma manual

```python
img = plt.imread('Imgs/cameraman.tiff')

num_bins = 50
bin_size = 256 / num_bins

# CÁLCULO DE HISTOGRAMA MANUAL
hist = np.zeros(num_bins)
num_rows, num_cols = img.shape
for row in range(num_rows):
  for col in range(num_cols):
    bin_index = int(img[row, col] / bin_size) 
    hist[bin_index] += 1
    
bins_values = np.linspace(0, 256, num_bins)

print('Size of each bin: %f'%bin_size)
print('Bin values: %s'%bins_values)
print('Histogram:')
print(hist)

plt.bar(bins_values, hist, bin_size, edgecolor = 'k')
```

##### Cálculo de histograma com numpy

```python
hist, bins = np.histogram(img.flatten(), num_bins)
```

```python
plt.figure(figsize = (12, 4))
plt.subplot(1, 2, 1) # linha de plots, colunas de plot, index do plot
plt.imshow(img, 'gray')
plt.subplot(1, 2, 2)
_ = plt.hist(img.flatten(), num_bins, edgecolor='k')
```

```python
# Aplica transformação (correção gamma) 
# para tornar a imagem mais escura, 
# e recalcula o histograma
gamma = 4.
img_transf = img ** gamma

plt.figure(figsize = (12, 4))
plt.subplot(1, 2, 1) # linha de plots, colunas de plot, index do plot
plt.imshow(img_transf, 'gray')
plt.subplot(1, 2, 2)
_ = plt.hist(img_transf.flatten(), num_bins)
```

```python
gamma = 0.5
img_transf = img ** gamma

plt.figure(figsize = (12, 4))
plt.subplot(1, 2, 1) # linha de plots, colunas de plot, index do plot
plt.imshow(img_transf, 'gray')
plt.subplot(1, 2, 2)
_ = plt.hist(img_transf.flatten(), num_bins)
```
